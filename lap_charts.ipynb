```python
import tabula
import pandas as pd
import os
```

```python
ref_df = pd.read_csv('which_race.csv', header=0)
pdf_list = os.listdir('raw_data')
```

```python
d5={}
for _pdf in pdf_list:
    if ('lap' in _pdf)&('chart' in _pdf):
        year = int(_pdf[:4])
        race_in_year = int(_pdf[5:].split('_')[0])
        race_id = ref_df['race_id'][(ref_df['year']==year)&(ref_df['race_in_year']==race_in_year)].values[0]
#         if (race_id!=7) & (race_id!=20)& (race_id!=21)& (race_id!=25)& (race_id!=26):
        if race_id not in [7,20,21,25,26]:
            d5['lap_chart_df_'+str(race_id)] = tabula.read_pdf('raw_data/'+_pdf,pages='all')
        elif race_id in [7,20,21,25,26]:
            df = tabula.read_pdf('raw_data/'+_pdf,pages=1,pandas_options = {'error_bad_lines':False})
            df2 = tabula.read_pdf('raw_data/'+_pdf,pages=2)
            d5['lap_chart_df_'+str(race_id)] = pd.concat([df,df2],axis=0, sort=False)
            #
#         elif (race_id==7)  (race_id==20)& (race_id!=21)& (race_id!=25)& (race_id!=26):
```
